```python
data_out = json.load(open("sample_files/sample_output.json", "r"))
data_out
```

```python
class gantt_chart(object):
    def __init__(self, data_path, data_type):
        data = json.load(open(data_path, "r"))
        if data_type == "input":
            self.process_df_in(data)
        elif data_type == "output":
            self.process_df_out(data)
        
        
    def process_df_in(self, data):
        proposals = {}
        for p in data["proposals_by_id"]:
            proposals[p] = int(data["proposals_by_id"][p]["tac_priority"])
        print(proposals)
        
        request_list = []
        for rg in data["json_request_group_list"]:
            proposal = rg["proposal"]
            proposal_priority = proposals[proposal]
            priority = proposal_priority * rg["ipp_value"]
            for r in rg["requests"]:
                r_id = r["id"]
                
                telescope = r["location"]["telescope_class"]
                for w in r["windows"]:
                    window = {
                        "telescope": telescope,
                        "start": dt.strptime(w["start"], "%Y-%m-%dT%H:%M:%SZ"),
                        "end": dt.strptime(w["end"], "%Y-%m-%dT%H:%M:%SZ"),
                        "id": r_id,
                        "priority": priority,
                        "proposal": proposal
                    }
                    request_list.append(window)
        df = pd.DataFrame(request_list)
        
        chart_start = df["start"].min()
        chart_end = df["end"].max()
        
        df["delay"] = (chart_start - df["start"]) / pd.Timedelta(minutes=1)
        df["duration"] = (df["end"] - df["start"]) / pd.Timedelta(minutes=1)
        
        print(df.head())
        
        fig, ax = plt.subplots(1, figsize=(16,6))
        
        ax.barh(df.id, df.duration, left=df.delay)
        
        
    
    def process_df_out(self, data):
        schedule_start = dt.strptime(data["schedule_start"], "%Y-%m-%dT%H:%M:%S.%f")
        schedule_end = dt.strptime(data["schedule_end"], "%Y-%m-%dT%H:%M:%S.%f")
        
        telescopes = list(data["resources"].keys())
        reservation_list = []
        for tel in telescopes:
            for reservation in data_out["resources"][tel]["reservations"]:
                reservation_list.append(reservation)
        df = pd.DataFrame(reservation_list)
        df["start"] = pd.to_datetime(df["start"], format="%Y-%m-%dT%H:%M:%S")
        df["end"] = pd.to_datetime(df["end"], format="%Y-%m-%dT%H:%M:%S")
        df["duration"] = (df["end"] - df["start"]) / pd.Timedelta(minutes=1)/3600
        df["delay"] = (df["start"] - schedule_start) / pd.Timedelta(minutes=1)/3600
        
        print(df.head())
        
        fig, ax = plt.subplots(1, figsize=(16,6))
        
        ax.barh(df.resource, df.duration, left=df.delay)
```

```python
gantt_chart("sample_files/sample_output.json", "output")
```
